### Collabrative Filtering
- Similar book prediction based on user feedback 

```python
books_with_ratings.head()
```

```python
## Grouping based on user-id cells the no of books rated by each user
x = books_with_ratings.groupby('User-ID').count()
x
```

```python
y
```

```python
y.reset_index()
```

```python
# The above dataframe tell mw how may users read the book.
y.sort_values('User-ID',ascending=False)
```

```python
# Now want user count greate then 50
y = y['Book-Rating'] >=50
```

```python
y
```

```python
famous_book = y[y].index
```

```python
famous_book
```

```python
final_ratings = filtered_data[filtered_data['Book-Title'].isin(famous_book)]
```

```python
final_ratings
```

```python
# Pivot table giving the rating for rach book for each user
# book row with userid as column
pt = final_ratings.pivot_table(index = 'Book-Title',columns = 'User-ID',values = 'Book-Rating')
```

```python
pt
```

```python
pt=pt.fillna(0)
```

```python
pt
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
type(similarity_score)
```

```python
df_temp =pd.DataFrame(similarity_score)
```

```python
df_temp
```

```python
pt.index
```

```python
df_temp_name = df_temp
```

```python
df_temp_name.index = pt.index
df_temp_name.columns = pt.index
```

```python
df_temp_name
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    # Lets create an empty list and in that list i want to popluate with the book information
    # Book-Title,Book-Authors,Image-url
    # Empty List
    data =[]
    for i in similar_item:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('1984')
```

```python
recommend('Animal Farm')
```
